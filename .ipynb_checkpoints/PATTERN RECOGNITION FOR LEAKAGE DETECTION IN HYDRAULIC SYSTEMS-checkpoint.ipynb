# Pattern Recognition for Leakage Detection in Hydraulic Systems

A Machine Learning pipeline composed of a [`Kalman Filter`](https://en.wikipedia.org/wiki/Kalman_filter), an `Oriented Gradient` algorithm and a `Pattern Classifier` which is based on [`Hierarchical Clustering Analysis`](https://en.wikipedia.org/wiki/Hierarchical_clustering) is proposed.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
from KF1D_Gradient import AutoRegressor

%matplotlib inline
```

```python
# Loading data
df = pd.read_parquet('./Data/HP-TankLevel 27JUN-08JUL_Leakage 2023.07.01.parquet')
```

# Classification Model

![Model](./Static/ML-PATTERN-CLASSIFIER.png)

## Resolution Amplifier (Factor)

To improve the resolution a factor of 10 was selected.

```python
d10 = df.copy()
```

## Estimation and Gradient Calculation

```python
# MODEL SETTINGS: E-MEASUREMENT: 19.0 ; E-PROCESS: 0.5 ; RESOLUTION FACTOR: 10
kf3 = AutoRegressor(19.0, 0.5, 1)
data = d10['TankLevel'].values
es3 = []
roc3 = pd.DataFrame()

for z in data:
    est = kf3.estimate(z)
    es3.append(est)
    grd = kf3.gradient()
    if grd:
        dft = pd.DataFrame([grd])
        roc3 = pd.concat([roc3, dft], ignore_index=True)
        roc3.reset_index()
```

```python
dates = df.index.values
fig, ax = plt.subplots(1, figsize=(15,15))
ax.plot(dates, data, linestyle='solid', linewidth=1.1, color='dodgerblue')
ax.plot(dates, es3, linestyle='solid', linewidth=1.1, color='darkorange')

ax.set_title("Hydraulic Tank Level", fontsize=18)
ax.set_ylabel("Inches/10", fontsize=16)
ax.grid()
```

## Oriented Gradients Distribution

```python
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.scatter(roc3['ang'].values, roc3['mag'].values)

ax.set_title("Phase and Magnitud", fontsize=18)
ax.set_xlabel("Gradient Phase", fontsize=16)
ax.set_ylabel("Gradient Magnitud", fontsize=16)

ax.grid()
plt.show()
```

## Dendrogram Construction

```python
X = roc3[['ang','mag']].values
linkage_matrix = linkage(X, method='ward')
```

```python
plt.figure(figsize=(20,15))
plt.title("Dendrogram", fontsize=22)
dendogram = dendrogram(linkage_matrix)
plt.show()
```

## Oriented Gradients Classification Through Hierarchical Clustering

```python
clp3 = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='ward')
Xp3 = roc3[['ang','mag']].values
clp3.fit_predict(Xp3)
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.scatter(Xp3[:,0], Xp3[:,1], c=clp3.labels_, cmap='rainbow')

ax.set_title("Phase and Magnitud Clusters", fontsize=18)
ax.set_xlabel("Gradient Phase", fontsize=16)
ax.set_ylabel("Gradient Magnitud", fontsize=16)

ax.grid()
plt.show()
```

## Pattern Recognition According with Gradient Classification

```python
roc3['cluster'] = clp3.labels_
gp3 = roc3.groupby('cluster').median()
gp3.sort_values('ang')
```

```python
# Setting colors and markers for type of cluster (7)
cs3 = dict()
mk3 = dict()
for cluster, _ in gp3.iterrows():
    if cluster == 2:
        cs3[2] = "black"
        mk3[2] = ","
    elif cluster == 4:
        cs3[4] = "red"
        mk3[4] = "d"
    elif cluster == 5:
        cs3[5] = "orange"
        mk3[5] = "v"
    elif cluster == 0:
        cs3[0] = "cyan"
        mk3[0] = "1"
    elif cluster == 3:
        cs3[3] = "royalblue"
        mk3[3] = "2"
    elif cluster == 1:
        cs3[1] = "blueviolet"
        mk3[1] = "^"
    elif cluster == 6:
        cs3[6] = "magenta"
        mk3[6] = "D"
```

```python
dg3 = d10.copy()
dg3.drop(dg3.index[0], axis=0, inplace=True)
dg3.reset_index(inplace=True)
pp3 = pd.concat([dg3, roc3], axis =1)
```

```python
xax = pp3['Time']
typ = pp3['cluster']
estm = list(map(lambda x: x/10 ,es3[1:]))

fig, ax = plt.subplots(1,figsize=(16, 16))
ax.plot(xax, estm, linestyle='solid', linewidth=1, color='silver')
for i in range(pp3.shape[0]):
    c = cs3[typ[i]]
    k = mk3[typ[i]]
    ax.plot(xax[i], estm[i], color=c, marker=k)

ax.set_title("Pattern Detection", fontsize=18)
ax.set_ylabel("Inches", fontsize=16)
ax.grid()
plt.show()
```
